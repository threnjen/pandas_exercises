### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', delimiter="\t")
```

### Step 4. How many products cost more than $10.00?

```python
#change the price column into a float
chipo['item_price'] = chipo['item_price'].map(lambda x: x[1:]).astype(float)
```

```python
# change the price column into a per-item price (the data comes in as price for all quantities ordered)
chipo['item_price'] = round((chipo['item_price'] / chipo['quantity']), 2)
chipo.sort_values('item_price').head(20)
#ten = chipo.loc[chipo['item_price'] > 10.00]
#len(ten['item_name'].unique())
#ten['item_name'].value_counts()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.loc[:,('item_name', 'item_price')]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(['item_price', 'quantity'], ascending = False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl'].sum()['quantity']
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
sodas = chipo[(chipo['item_name'].str.contains('Canned')) & (chipo['quantity'] > 1)]
sodas
```
